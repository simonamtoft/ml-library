# Define Model

```python
# Model Parameters
x_dim   = 784
z_dim   = 32
h_dims  = [512, 256]
```

```python
# define model
model = VariationalAutoencoder([x_dim, z_dim, h_dims], as_beta=True)
model.to(device)
model
```

# Train Model

```python
# Define Optimizer
optimizer = torch.optim.Adagrad(model.parameters(), lr=1e-2)

# This is the β-VAE
from itertools import repeat
beta = repeat(4.0)

# import loss function
from losses import binary_cross_entropy as loss_function
```

```python
from torch.autograd import Variable

# Training loop
for epoch in range(250):
    model.train()
    total_loss = 0
    for u, _ in iter(train):
        u = Variable(u.float()).to(device)

        reconstruction = model(u)
        
        likelihood = -loss_function(reconstruction, u)
        elbo = likelihood - next(beta) * model.kld
        
        L = -torch.mean(elbo).to(device)

        L.backward()
        optimizer.step()
        optimizer.zero_grad()

        total_loss += L.data

    m = len(train)

    if epoch % 1 == 0:
        print(f"Epoch: {epoch}\tL: {total_loss/m:.2f}")
```

# Sample (MNIST)

```python
x, *_ = next(iter(train))
```

```python
# Above image encoded to z
z, _, _ = model.encoder(Variable(x.float()).to(device))
z = z.repeat(10, 1)

# Perturb the first dimension of the first element
# the second dimension of the second element, etc.
for i in range(1, 10):
    z[i, i-1] += 10

samples = model.sample(z)
```
